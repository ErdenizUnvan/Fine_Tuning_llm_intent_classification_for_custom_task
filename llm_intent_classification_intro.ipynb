```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("cartesinus/multilingual_minilm-amazon-massive-intent")
model = AutoModelForSequenceClassification.from_pretrained("cartesinus/multilingual_minilm-amazon-massive-intent")
```

```python
# Example text
text = "What's the weather like today?"

# Tokenize the input
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
```

```python
# Get the model's predictions
outputs = model(**inputs)

# Access logits (raw scores)
logits = outputs.logits

# Convert logits to probabilities
from torch.nn.functional import softmax
probs = softmax(logits, dim=1)

# Get the predicted label
predicted_label = probs.argmax(dim=1).item()
print(f"Predicted label ID: {predicted_label}")
```

```python

# Etiket sayısını kontrol et
num_labels = model.config.num_labels
print(f"Number of labels: {num_labels}")

# Eğer model açıklaması etiketleri sağlamıyorsa, id_to_label oluşturulmamış olabilir.
# Ancak bu numara ile kaç adet intent olduğunu öğrenebilirsiniz.
```

```python
from transformers import AutoConfig

# Modelin konfigürasyonunu yükle
config = AutoConfig.from_pretrained("cartesinus/multilingual_minilm-amazon-massive-intent")

# Etiket sayısını kontrol et
print(config.id2label)  # Etiket haritalamasını gösterir
```

```python
id_to_label=config.id2label
id_to_label
```

```python
text = "What's the weather like today?"

# Tokenize the input
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

# Get the model's predictions
outputs = model(**inputs)

# Access logits (raw scores)
logits = outputs.logits

# Convert logits to probabilities
from torch.nn.functional import softmax
probs = softmax(logits, dim=1)

# Get the predicted label
predicted_label = probs.argmax(dim=1).item()
print(f"Predicted label ID: {predicted_label}")
id_to_label=config.id2label
print(f'Prediction: {id_to_label[predicted_label]}')
```
